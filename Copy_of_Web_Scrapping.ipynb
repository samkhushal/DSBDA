```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link='https://www.amazon.in/Realme-Cyber-Black-128GB-Storage/product-reviews/B0915ZNCPR/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
page = requests.get(link)
```

```python
page
```

```python
page.content
```

```python
soup = bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('span',class_='a-profile-name')
```

```python
names
```

```python
cust_name = []
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
cust_name
```

```python
cust_name.pop()
cust_name
```

```python
cust_name.pop(2)
cust_name
```

```python
cust_name.pop(4)
cust_name
```

```python
cust_name.pop(1)
cust_name
```

```python
title = soup.find_all('a',class_='review-title-content')
```

```python
title
```

```python
review_title = []
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title
```

```python
review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

```python
review_title.pop(1)
```

```python
len(review_title)
```

```python
rating = soup.find_all('i',class_='review-rating')
rating
```

```python
rate = []
for i in range(0,len(rating)):
    rate.append(rating[i].get_text())
rate
```

```python
review = soup.find_all("span",{"data-hook":"review-body"})
review
```

```python
review_content = []
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content
```

```python
review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
review_content
```

```python
review_content.pop(0)
```

```python
review_content[:] = [reviews.rstrip('\n') for reviews in review_content]
review_content
```

```python
len(review_content)
```

```python
import pandas as pd

# Assuming df is your DataFrame and review_title, rate, and review_content are your lists
# Check the length of the DataFrame's index
df_index_length = len(df.index)

# Check the length of each list
review_title_length = len(review_title)
rate_length = len(rate)
review_content_length = len(review_content)

# Check if lengths match
if df_index_length == review_title_length == rate_length == review_content_length:
    # Assign lists to DataFrame columns
    df['Review title'] = review_title
    df['Ratings'] = rate
    df['Reviews'] = review_content
else:
    # If lengths don't match, print an error message or handle the mismatch accordingly
    print("Lengths of values do not match length of index.")
```
